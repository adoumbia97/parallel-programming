## Import the libraries

```python
import numpy as np
import pandas as pd
from joblib import Parallel, delayed
import multiprocessing
# For the plot
import matplotlib.pyplot as plt
import seaborn as sns
from multiprocessing import Process
from multiprocessing import Value, Array
```

```python
df=pd.read_csv("../../GlobalLandTemperaturesByCity.csv")
```

## Definiton of the functions which will be used by the threads

```python
def f(dff,start,stop, list_mean, list_max,i):
    df_mean=dff.iloc[start:stop].groupby("Country").mean()
    df_max=dff.iloc[start:stop].groupby("Country").max()
    list_mean.append(df_mean)
    list_max.append(df_max)
```

## Definition of the variables to control the start, stop and the stop of each thread

```python
N=df.shape[0]
n=8
start=0
step=int(N/n)
stop=step
```

## Multiprocessing section 
- Divide the dataset in 8 sub dataset
- We create 8 threads
- Each thread will compute a specific dataset

```python
 if __name__ == '__main__':
    manager = multiprocessing.Manager()
    list_mean = manager.list()
    list_max=manager.list()
    jobs = []
    for i in range(n):
        p = Process(target=f, args=(df,start,stop,list_mean,list_max, i),name=i)
        jobs.append(p)
        p.start()
        
        start=stop
        stop=stop + step
        
    for proc in jobs:
        proc.join()  
        
```

- Combine the result of each thread 
- Compute the meand and max

```python
df_mean=pd.concat(list_mean)
df_mean=df_mean.groupby(df_mean.index).mean().sort_values("AverageTemperature", ascending=False)
df_max=pd.concat(list_max)
df_max=df_max[["AverageTemperature", "City", "dt"]].groupby(df_max.index).max().sort_values("AverageTemperature", ascending=False)
```

### The plotting of the average temperature per country and the maximum temperature by country

##### a. Average temperature per country 

```python
y = df_mean['AverageTemperature'].sort_values(ascending=True)
#Plotting
ax = y.plot(kind='barh', figsize=(20, 120), color='#df6d6d', zorder=2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
# Set x-axis label

ax.set_title("Average Temperature ", weight='bold', size=15)
# Set y-axis label
ax.set_ylabel("Country", weight='bold', size=12)
for i, v in enumerate(y):
    ax.text(v + 10, i, str(v), color='black', fontsize=10, ha='left', va='center', size=15)
    
plt.margins(0)
ax.margins(0)
plt.tight_layout()
```

##### b. Maximum temperature per country 

```python
y = df_max['AverageTemperature'].sort_values(ascending=True)
#Plotting
ax = y.plot(kind='barh', figsize=(20, 120), color='#df6d6d', zorder=2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
# Set x-axis label
#ax.set_xlabel("Number of fires", labelpad=20, weight='bold', size=12)
ax.set_title("Maximum Temperature ", weight='bold', size=15)
# Set y-axis label
ax.set_ylabel("Country", weight='bold', size=12)
for i, v in enumerate(y):
    ax.text(v + 10, i, str(v), color='black', fontsize=10, ha='left', va='center', size=15)
    
plt.margins(0)
ax.margins(0)
plt.tight_layout()
```
